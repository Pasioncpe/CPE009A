```python
# question 23
# Write a program that uses nested loops to print this pattern:

# *
# **
# ***
# ****
# *****
# ****
# ***
# **
# *

i = 0
while i != 10:
  if i < 5:
    i += 1
    for count in range(i):
      print("*", end='')
    print("")
  else:
    i += 1
    for count in range(10-i):
      print("*", end='')
    print("")



```
