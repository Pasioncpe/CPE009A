Create a class of Shapes (triangle, circle, square, rectangle), name properties (name and sides), method (function that compute for the area of each shape)

```python
#class for shapes including name, lenght of sides, function for area, and
class Shapes:
  def __init__(self, name, sides):
    self.name = name
    self.sides = sides

#function to get area exclusively for square, rectangle, triangle, and circle
  def area(self):
    if self.name == "square":
      return self.sides ** 2
    elif self.name == "rectangle":
      return self.sides[0] * self.sides[1]
    elif self.name == "triangle":
      return (self.sides[0] * self.sides[1]) / 2
    elif self.name == "circle":
      return 3.14 * (self.sides ** 2)

#function to print out description exclusively for
  def description(self):
    if self.name == "square":
      return f"A {self.name} with side length {self.sides} has an area of {self.area()}"
    elif self.name == "circle":
      return f"A {self.name} with radius {self.sides} has an area of {self.area()}"
    elif self.name == "rectangle":
      return f"A {self.name} with width {self.sides[0]} and length {self.sides[1]} has an area of {self.area()}"
    elif self.name == "triangle":
      return f"A {self.name} with base {self.sides[0]} and height {self.sides[1]} has an area of {self.area()}"
    else:
      return f"A {self.name} with properties {self.sides} has an area of {self.area()}"

#instances for the shapes
square = Shapes("square", 5)
rectangle = Shapes("rectangle", [4, 6])
triangle = Shapes("triangle", [3, 4])
circle = Shapes("circle", 7)

print(square.description())
print(rectangle.description())
print(triangle.description())
print(circle.description())
```
