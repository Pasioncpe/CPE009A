FOR LOOP

```python
#example 1

week =["Sunday","Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
for x in week:
  print(x)
```

```python
#example 2
week =["Sunday","Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
for x in week:
  print(x)
  if x == "Thursday":
    break
```

```python
#example 3
week =["Sunday","Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
for x in week:
  if x == "Thursday":
    break
  print(x)
```

```python
#example 4
for x in "week":
  print(x)
```

RANGE FUNCTION

```python
#example 1
for x in range(6):
  print(x)

print("\n")
for x in range(2,6):
  print(x)

#example 1.2
print("\n")
for x in range(2,6,2):
  print(x)
```

NESTED LOOPS

```python
#example 1
adjective = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adjective:
  for y in fruits:
    print(x,y)
```

WHILE LOOP

```python
#example 1
i = 1
while i < 6:
  print(i)
  i += 1
```

BREAK STAEMENT

```python
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

CONTINUE STATEMENT

```python
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```

ELSE

```python
i = 1
while i<6:
  print(i)
  i+=1
else:
  print("i is no longer less than 6")
```
