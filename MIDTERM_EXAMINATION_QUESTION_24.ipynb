Write a function that checks if a string of parentheses (), brackets [], and braces {} is balanced.

Examples:

"({[]})" → True

"([)]" → False

"((()))" → True

"({)}" → False


```python
# question number 24

def checker(message):
  i = 0
  if len(message) % 2 == 0:
    if message.count("(") == message.count(")") and message.count("[") == message.count("]") and message.count("{") == message.count("}"):

      while i < len(message)//2:
        i += 1
        if message[0] == "(":
          if message[len(message)-1] == ")":
            return True

        elif message[0] == "[":
          if message[len(message)-1] == "]":
            return True

        elif message[0] == "{":
          if message[len(message)-1] == "}":
            return True
  else:
    return False


while True:
  user = input("Please enter a message: ")

  if checker(user) == True:
    print("It is balanced: True")
  elif user == "done":
    break
  else:
    print("It is not balanced: False")
```
