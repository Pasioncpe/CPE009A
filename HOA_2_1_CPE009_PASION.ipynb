**Procedure:**

A teacher wants to calculate the final grade in a CpE course and want to write it in a python program. The following are the requirements:

  1. PRELIM GRADE = 50% Prelim Exam + 50% Prelim Class Standing (CS)  
  2. PRELIM CS = 50% Hands-on activity + 30% Quiz + 20% Assignment  
  3. MIDTERM GRADE = 1/3 of PRELIM GRADE + 2/3 of (50% Midterm Exam + 50% Midterm Class Standing (CS))  
  4. MIDTERM CS = 50% Hands-on activity + 30% Quiz + 20% Assignment  
  5. FINAL GRADE = 1/3 of MIDTERM GRADE + 2/3 of (50% Final Exam + 50% Final Class Standing (CS))  
  6. FINAL CS = 50% Hands-on activity + 30% Quiz + 20% Assignment  
  7. HOAs, Quizzes and Assignments are inputted as average of all submissions and are out of 100%.  
  8. Major exams are inputted out of 100%.  
  9. All computed grades are weighted according to the formulas above.  
  10. Final output shows PRELIM, MIDTERM, and FINAL grades clearly.  



```python
#Getting of class standing in PRELIM
pHOA = float(input("Enter Prelim HOA grade: "))
pQUIZ = float(input("Enter Prelim  Quiz grade: "))
pA = float(input("Enter Prelim Assessment Task grade: "))

pCS = pHOA*0.5 + pQUIZ* 0.3 + pA*0.2
print("Prelim Class standing is1: ", pCS)

#Getting of exam in PRELIM
pEX = float(input("Enter Prelim Exam grade: "))
print("Prelim Exam is: ", pEX)

#Getting of grade in PRELIM
pG = pEX * .5 + pCS * .5
print("Prelim Grade is: ", pG)

#Getting of class standing in MIDTERM
mHOA = float(input("Enter MidTerm HOA grade: "))
mQUIZ = float(input("Enter MidTerm  Quiz grade: "))
mA = float(input("Enter MidTerm Assessment Task grade: "))

mCS = mHOA*0.5 + mQUIZ* 0.3 + mA*0.2
print("MidTerm Class standing is: ", mCS)

#Getting of exam in MIDTERM
mEX = float(input("Enter MidTerm Exam grade: "))
print("MidTerm Exam is: ", mEX)

#Getting of grade in MIDTERM
mG = 1/3 * pG + 2/3 * (mEX*.5 + mCS*.5)
print("MidTerm Grade is: ", mG)

#Getting of class standing in FINAL
fHOA = float(input("Enter Final HOA grade: "))
fQUIZ = float(input("Enter Final  Quiz grade: "))
fA = float(input("Enter Final Assessment Task grade: "))

fCS = fHOA*0.5 + fQUIZ* 0.3 + fA*0.2
print("Final Class standing is: ", fCS)

#Getting of exam in FINAL
fEX = float(input("Enter Final Exam grade: "))
print("Final Exam is: ", fEX)

#Getting of grade in FINAL
fG = 1/3 * mG + 2/3 * (fEX*.5 + fCS*.5)
print("Final Grade is: ", fG)
```
