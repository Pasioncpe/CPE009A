1\. Make a program that would calculate the distance between two points given a list of coordinates. Use the distance formula.

```python
# created a function to make calculations faster
def disform(coordinates):
  x1 = coordinates[0][0]
  x2 = coordinates[0][1]
  y1 = coordinates[1][0]
  y2 = coordinates[1][1]
  distance = ((x2-x1)**2 + (y2-y1)**2)**0.5
  return distance

# given coordinates
coords = [(1,1), (2,3)]

# printing of distance
print(disform(coords))
```
