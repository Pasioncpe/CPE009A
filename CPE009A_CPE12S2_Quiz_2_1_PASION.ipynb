```python
# We want to create a python code for a grading scheme

# Create a function to accept a number of activities
# Compute for the score vs total
# Return the average grade of the student in %

def termCrit(actNumber):
  rawScore = []
  totalScore = []
  counter = 0

  for i in range(0, actNumber):
    print("\nActivity", i+1)
    inputScore = float(input("Score: "))
    inputTotal = float(input("Total: "))

    while counter != -1:
      if inputScore > inputTotal:
        print("\nInvalid Scores. Try again.\n")
        inputScore, inputTotal = 0,0
        inputScore = float(input("Score: "))
        inputTotal = float(input("Total: "))

      else:
        counter = -1

    rawScore.append(inputScore)
    totalScore.append(inputTotal)

  grade = sum(rawScore) / sum(totalScore)
  grade = grade * 100
  return grade
```

```python
# Create a function for the class standing per term
def termCS():
  print("\n{Computing for class standing}\n")
  print("{Hands-on Activities}")
  termHOA = termCrit(int(input("Number of Activities this term: ")))

  print("\n{Quizzes}")
  termQuiz = termCrit(int(input("Number of Activities this term: ")))

  print("\n{Assignment}")
  termAss = termCrit(int(input("Number of Activities this term: ")))

  grade = (0.5 * termHOA) + (0.3 * termQuiz) + (0.2 * termAss)
  return grade
```

```python
# Create a function to compute the term grade


def termGrade():
  print("{Computing for the term grade}\n")
  examRaw = float(input("Exam Score: "))
  examTotal = float(input("Exam Total: "))

  counter = 0

  while counter != -1:
    if examRaw > examTotal:
      print("\nInvalid Scores. Try again.\n")
      examRaw, examTotal = 0,0
      examRaw = float(input("Exam Score: "))
      examTotal = float(input("Exam Total: "))
    else:
      counter = -1

  examGrade = (examRaw/examTotal) * 100

  CSgrade = termCS()

  grade = (0.5 * examGrade) + (0.5 * CSgrade)
  return grade
```

```python
# Create a function to compute for the grades in the semester
def semesterGrade():
  print("---Prelim Grade---\n")
  prelimGrade = termGrade()

  print("\n---Midterm Grade---\n")
  midtermGrade = termGrade()
  finalMG = (1/3)*prelimGrade + (2/3)*midtermGrade

  print("\n---Final Grade---\n")
  finalGrade = termGrade()
  finalFG = (1/3)*finalMG + (2/3)*finalGrade

  return finalFG
```

```python
def convertGrade(finalGrade):
  if finalGrade >= 99.0: return 1.0
  elif finalGrade >= 96: return 1.25
  elif finalGrade >= 93: return 1.5
  elif finalGrade >= 90: return 1.75
  elif finalGrade >= 87: return 2.0
  elif finalGrade >= 84: return 2.25
  elif finalGrade >= 81: return 2.5
  elif finalGrade >= 78: return 2.75
  elif finalGrade >= 75: return 3.00
  else: return 0.0
  #The class should add everything else
```

```python
#Driver Code

# Trial 1
raw_final_grade = semesterGrade()
print("\n\nYour raw final grade percentage is:", raw_final_grade)
print("Your converted final grade is:", convertGrade(raw_final_grade), "\n\n 3 and above: Passing\n Below 3 or 10: Failed")
```
