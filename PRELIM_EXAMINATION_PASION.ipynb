```python
#Lists for the Subjects, Grades, and Remarks
subjects = ["Math", "Science", "English", "History"]
marks = []
finalgrade = []
finalmark = []


#Function to get mark and assign the grade
def calculation(i):

  grade = float(input(f"Grade of the Subject {subjects[i]}:"))
  marks.append(grade)
  if grade <= 100 and grade >= 90:
        return("A+")
  elif grade <= 89 and grade >= 80:
        return("A")
  elif grade <= 79 and grade >= 70:
        return("B+")
  elif grade <= 69 and grade >= 60:
        return("B")
  elif grade <= 59 and grade >= 50:
        return("C+")
  elif grade <= 49 and grade >= 40:
        return("C")
  else:
        return("F")

#Function to get the mark and assign remarks
def remark(grade):
  if grade >= 90:
    return("Outstanding")
  elif grade >= 80:
    return("Excellent")
  elif grade >= 70:
    return("Very Good")
  elif grade >= 60:
    return("Good")
  elif grade >= 40:
    return("Satisfactory")
  else:
    return("Needs Improvement")

#To run and append all values to lists
for i in range(0, len(subjects)):
  finalgrade.append(calculation(i))
  finalmark.append(remark(marks[i]))

#To print out grade, subject, and remark with the determined pass or fail
for i in range(0, len(subjects)):
  print("\n",subjects[i], ": ", finalgrade[i], ": ", finalmark[i])
  if marks[i] <= 40:
    print("FAILED")
  else:
    print("PASSED")
```
