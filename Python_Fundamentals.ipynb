Arithmetic Operations

```python
x = 5
y = 3

#Addition
print(x+y)

#Substraction
print(x-y)

#Multiplication
print(x*y)

#Division
print(x/y)

#Modulus
print(x%y)

#Exponentiation
print(x**y)

#Floor Derivative
print(x//y)
```

Assignment Operators

```python
# =
s = 1
print(s)
# +=
s = 1
s += 3
print(s)
# -=
s = 1
s -= 3
print(s)
# *=
s = 1
s *= 3
print(s)
# /=
s = 1
s /= 3
print(s)
# %=
s = 1
s %= 3
print(s)
# //=
s = 1
s //= 3
print(s)
# **=
s = 1
s **= 3
print(s)
# &=
s = 1
s &= 3
print(s)
# |=
s = 1
s |= 3
print(s)
# ^=
s = 1
s ^= 3
print(s)
# >>=
s = 1
s >>= 3
print(s)
# <<=
s = 1
s <<= 3
print(s)
```

Comparison Operators

```python
t = 1
u = 2

# ==
print(t == u)

# !=
print(t != u)

# >
print(t > u)

# <
print(t < u)

# >=
print(t >= u)

# <=
print(t <= u)
```

Logical Operators

```python
t = 1

print(t < 5 and t < 10)
print(t < 5 or t < 10)
print(not(t < 5 and t < 10))
```

Identity Operators

```python
t = 1
u = 5
print(t is u)
print(t is not u)
```

Bitwise Operators

```python
v = 6
# 2 = 0110

print(v & 1)
# v = 0000

print(v | 1)
# v = 0111

print(~v)
# v = 1001

print(v << 1)
# v = 1100

print(v >> 1)
# v = 0011
```
